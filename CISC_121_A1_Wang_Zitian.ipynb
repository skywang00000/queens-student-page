```python
def two_sum(x, y):
  for i in x:
    a = []
    a.append(i)
    for j in x:
      if i + j == y and i != j:
        a.append(j)
        return a
x = [1,2,3,4,5,6]
y = 7
b = two_sum(x,y)
print(b)
#output is [3,4]
```

# Problem 3:Longest Common Prefix

# 1.Understanding and breakdown
This question ask us to find the same pre-letters in the string. My idea: first ensure if it is a string. Then use for loop to chect each word step by step. Use the first word to compare with the rest. And compare the last letter one by one. If they are different, the length should -1 until same.

# 2.Peer review
I read someone's code:https://leetcode.com/problems/longest-common-prefix/solutions/7208886/optimal-python-solution-for-beginnners-with-o-n-m-and-o-m But his code is to complex,and it will affect the speed. I also asked AI, AI's code is as same as someone's I mentioned.

```python
def lcp(strs: list[str]):
    pre = strs[0]
    pre_len = len(pre)

    for s in strs[1:]:
        while pre != s[0:pre_len]:
            pre_len -= 1
            if pre_len == 0:
                return ""
            pre = pre[0:pre_len]
    return pre
strs = ["flower","flow","flight"]
print(lcp(strs))
```

# Problem 4: Move Zeros


# 1. Undersatanding and breakdown
This question asks me to find if there's zero in a list, if it is, move zero to the end. My initial idea is create a loop to find zero and delete it, and add zero in the end.

# 2.Peer Review
I view a solution from Leedcode:https://leetcode.com/problems/move-zeroes/solutions/72011/simple-o-n-java-solution-using-insert-index Initially it is complex. Now it becomes easier: we can use for loop to check where the position of list is equal to 0 and delete it. Eventually add the amount of 0 we deleted in the end.


```python
def zero(n: list[int]):
    n_len = len(n)
    for i in range(n_len):
        if n[i] == 0:
            del n[i]
            n.append(0)
    return n
n = [0,1,0,19,3,0,4,5]
print(zero(n))
```

# 3.Final Correct Code

```python
def zero(n: list[int]) -> list[int]:
    n_len = len(n)
    for i in range(n_len):
        if n[i] == 0:
            del n[i]
            n.append(0)
    return n
n = [0,1,0,19,3,0,4,5]
print(zero(n))
```
